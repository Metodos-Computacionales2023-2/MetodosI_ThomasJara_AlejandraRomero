Implemente el algoritmo de Gaus-Seidel para resolver el sistema de ecuaciones visto en
clase.
\
EL m´etodo de Gauss-Seidel se basa en el m´etodo de Jacobi con la ´unica diferencia que cada
variable se actualiza con cada iteraci´on interna. Usando el ejemplo anterior se evidencia como
funciona:

```python
def gauss_seidel(A,b,x0,tolerancia=1e-14,itmax=1000):
    x = x0.copy()
    sumk=np.zeros(len(x))
    residuo = np.linalg.norm( np.dot(A,x) - b)
    
    it = 0
    
    while residuo >= tolerancia and it < itmax:
        
        sumk[:]=0
        for i in range(A.shape[0]):
            for j in range(A.shape[1]):
                if i != j:
                    sumk[i] += A[i,j]*x[j]
            #print(sumk)
        
            x[i] = (b[i] - sumk[i])/A[i,i]
        #print(x)
        
        
        #print(x,residuo)
        # Norma infinita
        #residuo = np.max(np.abs(np.dot(A,x) - b))
        # Norma griega
        residuo = np.linalg.norm( np.dot(A,x) - b)
        
        it += 1
        
        if residuo > 1000:
            print('No convergio')
            x[:] = 0.
            break
        
    return x,it
A = np.array([[4, 1, 2],
              [3, 5, 1],
              [1, 1, 3]])
b = np.array([4, 7, 3])
x0 = np.zeros(len(b))
gauss_seidel(A,b,x0)
```
