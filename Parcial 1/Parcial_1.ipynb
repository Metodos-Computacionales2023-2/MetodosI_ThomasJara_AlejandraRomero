```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Funtion(x):
    return np.e**(-x)-x
```

```python
x = np.linspace(-1,10,50)
y = Funtion(x)
```

```python
plt.plot(x,y)
plt.axhline(y = 0,color='r')
```

```python
def cuadratica(f,x0,x1,x2,itmax=100,precision=1e-10):

    error = 1
    it = 0

    while error > precision and it < itmax:
        
        try:
            a=(((f(x2)-f(x1))/(x2-x1))-((f(x1)-f(x0))/(x1-x0)))/(x2-x0)
            b=((f(x1)-f(x0))/(x1-x0))-(x0+x1)*a
            c=f(x0)-x0*((f(x1)-f(x0))/(x1-x0))+x0*x1*a

            if b >= 0:
                x3 = (-2*c)/(b+np.sqrt((b**2)-4*a*c))
            else:
                x3 = (-2*c)/(b-np.sqrt((b**2)-4*a*c))

            error = np.abs(f(x3))
            
        except ZeroDivisionError:
            print('Division por cero')

        x0=x1
        x1=x2
        x2=x3

        it += 1

    return x2
```

```python
cuadratica(Funtion,-2,4,3)
```
