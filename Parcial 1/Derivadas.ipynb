```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def DerivadaDerechaO_2(f,x,h):
    
    d = 0.
    
    if h != 0:
        d = (1/(2*h))*((-3)*f(x) + 4*f(x+h) - f(x+2*h))
        
    return d
```

```python
def DerivadaCentral(f,x,h):
    
    d = 0.
    
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
        
    return d
```

```python
def Function(x):
    return np.sqrt(np.tan(x))
```

```python
def ExactDerivative(x):
    return (1/(np.cos(x))**2)*(1/(2*np.sqrt(np.tan(x))))
```

```python
N = 50
x = np.linspace(0.1,1.1,N)
h = 0.01
```

```python
PDerivative = DerivadaDerechaO_2(Function, x, h)
CDerivative = DerivadaCentral(Function,x,h)
EDerivative = ExactDerivative(x)
```

```python
plt.scatter(x,EDerivative,label='Exacta')
plt.scatter(x,PDerivative,label='Progresiva')
plt.scatter(x,CDerivative,label='Central')
plt.legend()
```

```python
ErrorP = np.abs(EDerivative-PDerivative)
ErrorC = np.abs(EDerivative-CDerivative)
```

```python
plt.scatter(x,ErrorP)
plt.scatter(x,ErrorC)
```
