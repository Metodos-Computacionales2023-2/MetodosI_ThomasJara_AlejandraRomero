```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
x = np.linspace(-1,1,n+1)
y = np.linspace(-1,1,n+1)
```

```python
def function(x,y,R=1.):
    
    z = R**2 - x**2 - y**2
    #print(x,y,z)
    if z < 0.:
        return 0.
    else:
        return np.sqrt(z)
    
```

```python
F = np.vectorize(function)
```

```python
X,Y = np.meshgrid(x,y)
Z = F(X,Y)
```

```python
V=0
A=(x[1]-x[0])**2
for i in range(len(Z)-1):
    for j in range(len(y)-1):
        Prom=(Z[i,j]+Z[i+1,j]+Z[i,j+1]+Z[i+1,j+1])/4
        V+=Prom* A 
print(V)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z)
```
